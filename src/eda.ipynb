## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
df = pd.read_csv("data/raw/combined_reviews.csv")
print(df.head())
```

```python
# text length
df["length"] = df["text"].apply(len)

# plot
sns.histplot(data=df, x="length", hue="label", bins=50)
plt.title("Text Length Distribution")
plt.show()
```

```python
# sentence count
df["sentences"] = df["text"].apply(lambda x: len(re.split(r"[.!?]+", x)))

print(df.groupby("label")[["length", "sentences"]].describe())


# plot sentence count
sns.histplot(data=df, x="sentences", hue="label", bins=10)
plt.title("Sentence Count Distribution")
plt.show()
```
